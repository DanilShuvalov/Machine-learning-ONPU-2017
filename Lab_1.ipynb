1) Использование библиотеки NumPy:

```python
import numpy as np
i=30
X=np.random.randint(-20, 20, size=(i))
Y=np.random.randint(-20, 20, size=(i))
print "X mass:"
print X
print "Y mass:"
print Y

```

Создаем два случайных одномерных массива X, Y и выводим их на экран.

2) С помощью Matplotlib отрисовываем полученные координаты в виде  линии зеленого цвета.

```python
import matplotlib.pyplot as ln
ln.plot(X, Y, 'green')
ln.show()
```

Рисуем линию зеленого цвета используя метод plot для объекта ln.

3) Проводим сортировку пузырьком.

```python
n = 1 
while n < len(X):
     for j in range(len(X)-n):
          if X[j] > X[j+1]:
               X[j],X[j+1] = X[j+1],X[j]
     n += 1
print "X mass:"
print X
```

На каждом шаге цикла while (начиная с первого элемента) очередной элемент выставляется в нужную позицию.(По возрастанию)

4) Изображаем полученные координаты в виде точек красного цвета.
